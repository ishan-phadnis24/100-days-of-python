```python
#take a string and check if it is palindrome or not eg. (madam)(eye)(wow)(mom)(malayalam)

o1 = input()

reversed_o1 = o1[::-1]

if reversed_o1 == o1:
  print('palindrome')
else:
  print("NA")
```
