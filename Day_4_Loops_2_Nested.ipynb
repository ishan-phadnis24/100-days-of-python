```python
#Q- print = 1,3,7,13,21,31,43 using FOR loop

num = 1
diff=2

for i in range(7):
  print(num,end=" ")
  num+=diff
  diff+=2
```

```python
#Q- print = 1,3,7,13,21,31,43 using WHILE loop
num = 1
diff = 2
while num <= 43:
  print(num,end=" ")
  num+=diff
  diff +=2
```

```python
#Q- print = 1,3,7,13,21,31,43 using WHILE loop using count
num = 1
diff = 2
count = 0

while count < 7:
    print(num, end=" ")
    num += diff
    diff += 2
    count += 1
```

```python
#BREAK
for i in range(5):
  if i ==3:
    break#breaks the loop
  print(i,end=" ")
```

```python
#CONTINUE
for i in range(5):
  if i ==3:
    continue#skip specific iteration
  print(i,end=" ")
```

```python
#PASS
for i in range(5):
  if i ==3:
    pass#acts as a placeholder
  print(i,end=" ")
```

```python
N = int(input())

for i in range(1, N - 1):
    for j in range(1, i - 1):
        print(j, end=" ")
    print()
```

```python
'''Print a pattern consisting of N rows, where each row contains an asterisk (*) at the beginning and end of the line, with N-2 spaces in between.

The Pattern should look like:

*<N-2 Spaces>*

Print the above pattern for a total of N Rows.

'''

N = int(input())

for i in range(N):
    print("*" + " " * (N - 2) + "*")
```

```python
#GCB of 2 Numbers
A = int(input())
B = int(input())

X = min(A,B)

for i in range(X,0,-1):
  if A%i ==0 and B%i ==0:
    print(i)
    break
```

```python
#LCM of 2 Numbers
A = 6
B = 8

X = max(A, B)  # Start from the larger number

# Go upward checking multiples
for i in range(X, A * B + 1,1):
    if i % A == 0 and i % B == 0:
        print(i)
        break
```
