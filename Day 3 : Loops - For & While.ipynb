```python
#use while loop to print even numbers from start to end assuming start<end

start=int(input('Enter number:'))
end = int(input('Enter number:'))

while start<=end:
  if start%2==0:# Check if number is even
    print(start)
  start+=1
```

```python
#Q= calculate sum of all numbers from 1 to N(input) using a for loop
n = int(input("enter number"))
total = 0
for i in range(1,n):
  print(i)
  total+=i
print(f"total_sum:{total}")
```

```python
#Q= print sum of all numbers from 1 to n where n is user input using FOR loop
n = int(input("enter number:"))

total_sum = 0

for i in range(1,n+1):
  print(i)
  total_sum+=i
print("SUM:",total_sum)
```

```python
#Q= print average of sum all numbers from 1 to n where n is user input using FOR loop

n = int(input("enter number: "))
total = 0  # Initialize sum

for i in range(1, n+1):
    total += i  # Add each number to total

average = total / n  # Average = sum / count
print(f"Average: {average}")
```

```python
#Q:Take an input of a number A from the user. Print all perfect squares less than or equal to A.

A = int(input())


i = 1
while i * i <= A:
    print(i * i)
    i += 1
```
